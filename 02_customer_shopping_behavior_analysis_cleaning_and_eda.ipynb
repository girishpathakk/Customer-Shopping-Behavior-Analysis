```python
# Treatig null values or replacing all values in Electronics category 
df.loc[df['Category']=='Electronics', 'Size']="Not Applicable"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# Treatig null values or replacing all values in Accesories category 
df.loc[df['Category']=='Accessories', 'Size']="Free Size"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# Treatig null values or replacing all values in Footwear category 
df.loc[df['Category']=='Footwear', 'Size']="Not Available"
```

```python
# Treatig null values or replacing all values in Clothing category
df.loc[(df["Category"]=="Clothing") & (df["Size"].isnull()), "Size"] =df[df["Category"]=="Clothing"]["Size"].mode()[0]
```

### Review Rating

```python
df.head()
```

```python
#mean of review rating
df['Review Rating'].mean()
```

```python
#median of review rating
df['Review Rating'].median()
```

```python
#mode of review rating
df['Review Rating'].mode()
```

```python
df.groupby("Item Purchased")['Review Rating'].mean()
```

```python
# Filling null values of Review ratings with mean of each item purchased
df['Review Rating']=df[ 'Review Rating'].fillna(df.groupby ('Item Purchased')['Review Rating'] .transform ('mean'))
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

```python
df['Review Rating']=df['Review Rating'].round(2)
```

### Purchase Amount (USD)
